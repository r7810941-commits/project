```python
import numpy as np
products=["Mobile","Light","TV","Cycle","Bike"]
#sales units
sales=np.array([1000,200,8000,5000,9000])
#Total sales
print("Total sales:",np.sum(sales))
#Average sales
print("Average sales:",np.mean(sales))
#Highest sales
print("Highest sales:",np.max(sales))
#Lowest sales
print("Lowest sales:",np.min(sales))
#Best selling product
best_index=np.argmax(sales)
print("Best selling product:",products[best_index])
#product-wise sales
print("\n product sales report")
for i in range(len(products)):
    print(products[i],":",sales[i])
```

```python
import pandas as pd
data={
    'Product':['Laptop','mobile','None','Tablet','Laptop'],
    'Sales':[50000,30000,20000,None,50000],
    'Quantity':[5,3,None,2,5]
}
df=pd.DataFrame(data)
print("Original Dataset:")
print(df)
df=df.drop_duplicates()
df['Product']=df['Product'].fillna('unknown')
df['Sales']=df['Sales'].fillna(df['Sales'].mean())
df['Quantity']=df['Quantity'].fillna(0)
print("\nCleaned Dataset:")
print(df)
```

```python
region_sales=df.groupby("region")["sales"].sum()
print(region_sales)
```

```python
product_sales=df.groupby("product")["sales"].sum()
print(product_sales)
```

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May","Jun"]
sales=[10000,12000,15000,13000,18000,20000]
plt.figure(figsize=(12,5))# Adjusted figsize to accommodate three plots
#Line Chart
plt.subplot(1,2,1)# Charged to 1 row,3 columns,position 1
plt.plot(months,
        sales,
        marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
#Bar chart
plt.subplot(1,2,2)# Charged to 1 row,3 columns,position 2
plt.bar(months,
        sales,
        color="orange")
plt.title("Sales comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
#pie chart
plt.subplot(1,3,3)# Charged to 2 row,3 columns,position 3
expenses=[5000,3000,2000]
categories=["Food","Travel","Shopping"]
plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%",colors=["pink","yellow","green"])
plt.title("Expenses Distribution")
plt.tight_layout()
plt.show()
```
