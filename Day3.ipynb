```python
def attendance(attended,total):
  return (attended / total) *100
attended=int(input("classes attended:"))
total=int(input("total classes:"))
percent=attendance(attended,total)
print("attendance percentage=",round(percent,2),"%")
```

```python
def calculate_average(x,y,z):
  return(x+y+z)/3
try:
  x=int(input("enter first number:"))
  y=int(input("enter second number:"))
  z=int(input("enter third number:"))
  print("calculate_average=",calculate_average(x,y,z))
except valueError:
  print("invalid input")
```

```python
def attendance_percentage(attended,total):
    return(attended/total)*100
def check_eligibility(percentage):
    if percentage>=75:
       return"ELIGIBLE"
    else:
       return"NOT ELIGIBLE"
#Input
total=int(input("enter total classes:"))
attended=int(input("enter classes attended:"))
#Calculate
perc=attendance_percentage(attended,total)
#Display result with status
print(f"Attenndence:{perc:.2f}%")
print(f"Status:{check_eligibility(perc)}")
```
